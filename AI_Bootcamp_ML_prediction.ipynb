```python
import pandas as pd
import pickle
```

```python
# read my model using pickle lib
best_model=pickle.load(open('/svc_diab.pkl','rb'))
```

```python
# create random input data

input_data={"Pregnancies":2,
	"Glucose":210,
  "BloodPressure":90,
  "SkinThickness":25,
  "Insulin":240,
  "BMI":30,
  "DiabetesPedigreeFunction":2.5,
  "Age":35}
```

```python
input=pd.DataFrame(input_data,index=[0])
```

```python
result=best_model.predict(input)[0]

if result==1:
  print("The person is diabetic")
else:
  print("The person is not diabetic")
```
